# 单群

```python
import math

# 构造一个矩阵保存结果
m = 1001
l = 0
mat = []

while l<m:
    line = []
    mat.append(line)
    l = l+1

print(mat)
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n==2:
        return True
    if n%2==0:
        return False
    for i in range(3,int(math.isqrt(n))+1,2):
        if n%i==0:
            return False
    return True
```

```python
for i in range(1,1000):
    if is_prime(i):
        mat[i].append(solutions[1])
print(mat)
```
